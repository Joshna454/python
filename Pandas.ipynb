**Locate Row:** Pandas use loc attribute to return one or more specified rows.
**iloc** iloc property gets,or sets the values of a specified indexes.

```python
import pandas as pd
data = [100,102,202]

series = pd.Series(data, index=["a","b","c"])
series.loc["c"] = 200
print(series)
```

```python
#iloc  = interger location
import pandas as pd

values = [58,85,69,96,22]
series = pd.Series(values, index=["a","b","c","d","e"])
print(series.iloc[0])
print(series[series > 50])
```

```python
import pandas as pd

calories = {"Day1" : 1750, "Day2" : 2100, "Day3" : 1700}
series = pd.Series(calories)

series.loc["Day3"] += 400
print(series)
```

```python
**DataFrames**: A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.
```

```python
import pandas as pd

data = {
    "Name" : ["Spongebob", "Patrick", "Squidward"],
    "Age" :[30,35,50]
}
df = pd.DataFrame(data, index = ["Employee 1", "Employee 2", "Employee 3"])

# Add a new Column
df["Job"] = ["Cook","N/A","Cashier"]

# Add a new row 

new_row = pd.DataFrame([{"Name":"Sandy","Age": 28,"Job": "Engineer"},
                       {"Name":"Eugene","Age": 60,"Job": "Manager"}],
                     index=["Employee 4","Employee 5"] )
df = pd.concat([df, new_row])
print(df)
```

```python
import pandas as pd

data = {
    "Name" : ["Spongebob", "Patrick", "Squidward"],
    "Age" :[30,35,50]
}
df = pd.DataFrame(data, index = ["Employee 1", "Employee 2", "Employee 3"])
print(df.loc["Employee 1"])
print(df.iloc[2])
```

```python
#Filtering = Keeping the rows that match a condition
import pandas as pd

df = pd.read_csv("one_piece.csv")

# Filter all Captains
captains = df[df["Role"] == "Captain"]
print(captains)
```

```python
import pandas as pd
df = pd.read_csv("one_piece.csv")

# Captain AND from East Blue arc and Multiple Conditions (AND)
result = df[(df["Role"] == "Captain") & (df["Arc"] == "East Blue")]
print(result)
```

```python
import pandas as pd
df = pd.read_csv("one_piece.csv")

# Characters appearing after episode 500
#Filter by Episode (Numbers)
result = df[df["Episode"] > 500]
print(result)
```

```python
import pandas as pd
df = pd.read_csv("one_piece.csv")

# Characters from Wano OR Marineford
#Multiple Conditions (OR)
result = df[(df["Arc"] == "Wano") | (df["Arc"] == "Marineford")]
print(result)
```

**Aggregation** in pandas is all about summarizing your data—turning many rows into meaningful insights like totals, averages, counts, etc.

```python
import pandas as pd

df = pd.read_csv("one_piece.csv")

df["Episode"].mean()
```

```python
import pandas as pd
df = pd.read_csv("one_piece.csv")

df["Episode"].max()
```

```python
import pandas as pd
df = pd.read_csv("one_piece.csv")

df["Episode"].min()
```

```python
import pandas as pd
df = pd.read_csv("one_piece.csv")

df["Name"].count()
```

```python
#GroupBy + Aggregation
import pandas as pd
df = pd.read_csv("one_piece.csv")

#Count characters per Arc
df.groupby("Arc")["Name"].count()
```

```python
import pandas as pd
df = pd.read_csv("one_piece.csv")

#Average Episode per Arc
df.groupby("Arc")["Episode"].mean()
```

```python
#Multiple Aggregations Together
import pandas as pd
df = pd.read_csv("one_piece.csv")

df.groupby("Arc")["Episode"].agg(["count", "mean", "max"])
```

```python
#Aggregation on Multiple Columns

import pandas as pd
df = pd.read_csv("one_piece.csv")

df.groupby("Arc").agg({
    "Episode": ["mean", "max"],
    "Name": "count"
})
```

```python
#Reset Index (Cleaner Output)
import pandas as pd
df = pd.read_csv("one_piece.csv")

result = df.groupby("Arc")["Name"].count().reset_index()
print(result)
```

**DATA CLEANING** it’s all about fixing messy, incorrect, or incomplete data so your analysis becomes accurate.

```python
#Check Your Data First

import pandas as pd
df = pd.read_csv("one_piece.csv")

print(df.head())      # first few rows
print(df.info())      # structure + data types
print(df.isnull().sum())  # missing values
```

```python
import pandas as pd
df = pd.read_csv("one_piece.csv")

df = df.dropna()
```

```python
import pandas as pd
df = pd.read_csv("one_piece.csv")

df["Arc"] = df["Arc"].fillna("Unknown")
df["Episode"] = df["Episode"].fillna(0)
```

```python
import pandas as pd
df = pd.read_csv("one_piece.csv")

df = df.drop_duplicates(subset="Name")
```

```python
#Fix Column Names
import pandas as pd
df = pd.read_csv("one_piece.csv")

df.columns = df.columns.str.strip().str.lower()
```

```python
#Fix Text Data - Removes spaces & standardizes format
import pandas as pd
df = pd.read_csv("one_piece.csv")

df["Name"] = df["Name"].str.strip()
df["Arc"] = df["Arc"].str.title()
```

```python
#Convert Data Types- Sometimes numbers are read as strings → fix them
import pandas as pd
df = pd.read_csv("one_piece.csv")

df["Episode"] = df["Episode"].astype(int)
```

```python
#Rename Columns
import pandas as pd
df = pd.read_csv("one_piece.csv")

df.rename(columns={"Episode": "episode_no"}, inplace=True)
```

```python
#Replace Wrong Values
import pandas as pd
df = pd.read_csv("one_piece.csv")

df["Arc"] = df["Arc"].replace("wano", "Wano")
```
