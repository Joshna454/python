```python
#This creates a simple line graph.
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 40]

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

**Why use NumPy?**
NumPy helps generate numerical data quickly (arrays, ranges, math functions).
- np.linspace(0,10,100) creates 100 evenly spaced values
- This gives you smooth curves (very important in ML graphs)

```python
import numpy as np
import matplotlib.pyplot as plt

# create x values
x = np.linspace(0, 10, 100)

# create y values
y = np.sin(x)

# plot
plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("X values")
plt.ylabel("sin(x)")

plt.show()
```

**Matplotlib Markers**
In Matplotlib, markers are shapes like:

- circles ⚫
- squares ⬛
- triangles 🔺

They help you see individual data points clearly, instead of just a line.

**Marker	Description**
'o'	Circle	
'*'	Star	
'.'	Point	
','	Pixel	
'x'	X	
'X'	X (filled)	
'+'	Plus	
'P'	Plus (filled)	
's'	Square	
'D'	Diamond	
'd'	Diamond (thin)	
'p'	Pentagon	
'H'	Hexagon	
'h'	Hexagon	
'v'	Triangle Down	
'^'	Triangle Up	
'<'	Triangle Left	
'>'	Triangle Right	
'1'	Tri Down	
'2'	Tri Up	
'3'	Tri Left	
'4'	Tri Right	
'|'	Vline	
'_'	Hline

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y= [10,20,15,25]

plt.plot(x,y,marker = 'o', markersize= 10,
         markerfacecolor = "purple",markeredgecolor = "purple",
        linestyle = "dashed", linewidth=4 ,
        color = "lavender") #'o' means circle marker
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x =np.array([2023,2024,2025,2026])
y1= np.array([10,20,15,25])
y2 = np.array([17,23,38,5])
y3 = np.array([13,15,20,30])

line_style = dict(marker = "o", markersize= 10,
         markerfacecolor = "purple",markeredgecolor = "purple",
        linestyle = "solid", linewidth=4)

plt.plot(x,y1, color="#ff0000", **line_style)
plt.plot(x,y2, color="#00ff00", **line_style)
plt.plot(x,y3, color="#6666ff", **line_style)

plt.show()
```

```python
#Labels
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105 ])
y = np.array([240, 250, 260, 270, 280, 290 ])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

```python
#grid() - Helps make plots easier to read by adding refernce lines.

import matplotlib.pyplot as plt
import numpy as np

x = [1,2,3,4,5]
y = [5,10,15,20,25]

plt.grid(axis="y", linewidth=1,color = "lightgray",linestyle = "dashdot")
plt.plot(x,y)
plt.show()
```
